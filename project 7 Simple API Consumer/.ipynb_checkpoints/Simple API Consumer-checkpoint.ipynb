```python
import numpy as np, pandas as pd
import json
import requests 
```

```python
def fetch_json(URL):
    url = URL
    res = requests.get(url, timeout = 5)
    
    if res.status_code == 200:
        return res.text
    else:
        print("Please Enter a valid Nmae")
        return None
```

```python
def parse_pokemon_details(raw_txt):
    data = json.loads(raw_txt)
    
    types = [ty['type']['name'].title() for ty in data['types']]
   
    return {
        'Name' : data['name'].title(),
        'Pokedex No' : data['id'],
        'Types' : ', '.join(types),
        'Height' : data['height'] / 10,
        'Weight' : data['weight']/ 10 
        
    }
    
```

```python
def print_pokemon(data):
    print("Name :",data["Name"])
    print("Pokedex No :",data["Pokedex No"])
    print("Type :",data["Types"])
    print("Height : ",data["Height"], "Meter")
    print("Weight : ",data["Weight"], "Kg")
```

```python
def main():
    name = input("Enter Pokemon Name: ").strip()
    url = f"https://pokeapi.co/api/v2/pokemon/{name}"
        

    raw = fetch_json(url) 
        
    if raw:
        data = parse_pokemon_details(raw)  
        
    if raw:
        print_pokemon(data)           
        

```

```python
main()
```
